## Facebook

```python
facebook_news=pd.read_csv('/kaggle/working/facebook_news.csv')
facebook_vaccines_all_comments_scored=pd.read_csv('/kaggle/working/facebook_vaccines.csv')
facebook_brexit_all_comments_scored=pd.read_csv('/kaggle/working/facebook_brexit.csv')
```

```python
# Seleziona solo la colonna post_id e aggiungi la colonna topic
facebook_vaccines = facebook_vaccines_all_comments_scored[['post_id','comment_id']].copy()
facebook_vaccines['topic'] = 'Vaccines'

facebook_brexit= facebook_brexit_all_comments_scored[['post_id','comment_id']].copy()
facebook_brexit['topic'] = 'Brexit'

facebook_news = facebook_news[['post_id','comment_id']].copy()
facebook_news['topic'] = 'News'

# Concatenazione dei dataframe
facebook = pd.concat([facebook_vaccines, facebook_brexit, facebook_news])

# Salva il risultato come un nuovo dataframe
facebook.reset_index(drop=True, inplace=True)
```

```python
facebook
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcola il numero di commenti per ogni post per i dati di vaccines
conversation_size_vaccines = facebook_vaccines_all_comments_scored.groupby(['post_id'])['comment_id'].count().reset_index()
conversation_size_vaccines.columns = ['post_id', 'comment_count']

# Calcola il numero di commenti per ogni post per i dati di news
conversation_size_news = facebook_news.groupby(['post_id'])['comment_id'].count().reset_index()
conversation_size_news.columns = ['post_id', 'comment_count']

# Calcola il numero di commenti per ogni post per i dati di brexit
conversation_size_brexit = facebook_brexit_all_comments_scored.groupby(['post_id'])['comment_id'].count().reset_index()
conversation_size_brexit.columns = ['post_id', 'comment_count']

# Conta quanti post hanno un determinato numero di commenti per i dati di vaccines
post_count_vaccines = conversation_size_vaccines.groupby('comment_count').size().reset_index(name='post_count')

# Conta quanti post hanno un determinato numero di commenti per i dati di news
post_count_news = conversation_size_news.groupby('comment_count').size().reset_index(name='post_count')

# Conta quanti post hanno un determinato numero di commenti per i dati di brexit
post_count_brexit = conversation_size_brexit.groupby('comment_count').size().reset_index(name='post_count')

# Disegna lo scatter plot con scala logaritmica su entrambi gli assi
plt.figure(figsize=(12, 8))

# Scatter plot per i dati di vaccines
sns.scatterplot(data=post_count_vaccines, x='comment_count', y='post_count', alpha=0.5, label='Vaccines', color='steelblue', edgecolor=None)

# Scatter plot per i dati di news
sns.scatterplot(data=post_count_news, x='comment_count', y='post_count', alpha=0.5, label='News', color='dodgerblue', edgecolor=None)

# Scatter plot per i dati di brexit
sns.scatterplot(data=post_count_brexit, x='comment_count', y='post_count', alpha=0.5, label='Brexit', color='lightskyblue', edgecolor=None)

plt.xscale('log')
plt.yscale('log')
plt.title('Facebook - Scatter plot of comments per post (log-log scale)')
plt.xlabel('Number of comments')
plt.ylabel('Number of posts')

# Rimuovi la griglia
plt.grid(False)

plt.legend(title='Topic')
plt.show()
```

```python
youtube=pd.read_parquet('/kaggle/working/youtube.parquet' ,engine='fastparquet')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supponiamo che il dataset si chiami youtube_data e abbia le colonne 'post_id', 'comment_id', e 'topic'
youtube_data = youtube

# Calcola il numero di commenti per ogni post per ogni topic
conversation_size = youtube_data.groupby(['topic', 'post_id'])['comment_id'].count().reset_index()
conversation_size.columns = ['topic', 'post_id', 'comment_count']

# Conta quanti post hanno un determinato numero di commenti per ogni topic
post_count = conversation_size.groupby(['topic', 'comment_count']).size().reset_index(name='post_count')

# Disegna lo scatter plot con scala logaritmica su entrambi gli assi
plt.figure(figsize=(12, 8))

# Ottieni i topic unici
topics = post_count['topic'].unique()

# Definisci una palette di colori per i diversi topic
palette = sns.color_palette("Reds", len(topics))

# Disegna lo scatter plot per ogni topic
for i, topic in enumerate(topics):
    topic_data = post_count[post_count['topic'] == topic]
    sns.scatterplot(data=topic_data, x='comment_count', y='post_count', alpha=0.5, label=topic, color=palette[i], edgecolor=None)

plt.xscale('log')
plt.yscale('log')
plt.title('YouTube - Scatter plot of comments per post (log-log scale)')
plt.xlabel('Number of comments')
plt.ylabel('Number of posts')

# Rimuovi la griglia
plt.grid(False)

plt.legend(title='Topic')
plt.show()
```

# Figure 1A

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Suppose the datasets are named reddit_data, voat_data, youtube_data, and facebook_data
# Each with columns 'post_id', 'comment_id', and 'topic'
datasets = {
    'Reddit': reddit,
    'Voat': voat,
    'Gab': gab,
    'Facebook': facebook
}

# Set up the figure with 2 rows and 2 columns of subplots
fig, axes = plt.subplots(2, 2, figsize=(20, 16))

# Define the palette with different intensities for each social media platform and each topic
palette = {
    'Reddit': sns.light_palette('#5296dd', n_colors=4, reverse=True),
    'Voat': sns.light_palette( '#800080', n_colors=4, reverse=True),
    'Gab': sns.light_palette('#00c853', n_colors=1, reverse=True),
    'Facebook': sns.light_palette('#3b5998', n_colors=4, reverse=True)
}

# Loop through the datasets and plot for each
for ax, (name, data) in zip(axes.flatten(), datasets.items()):
    # Calculate the number of comments per post for each topic
    conversation_size = data.groupby(['topic', 'post_id'])['comment_id'].count().reset_index()
    conversation_size.columns = ['topic', 'post_id', 'comment_count']
    
    # Count how many posts have a certain number of comments for each topic
    post_count = conversation_size.groupby(['topic', 'comment_count']).size().reset_index(name='post_count')
    
    # Get unique topics
    topics = post_count['topic'].unique()
    
    # Draw scatter plot for each topic
    for i, topic in enumerate(topics):
        topic_data = post_count[post_count['topic'] == topic]
        sns.scatterplot(data=topic_data, x='comment_count', y='post_count', alpha=0.5, label=topic, color=palette[name][i % len(palette[name])], edgecolor=None, ax=ax)
    
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_title(f'{name} ')
    ax.set_xlabel('Number of comments')
    ax.set_ylabel('Number of posts')
    ax.grid(False)
    ax.legend(title='Topic')
    
    # Set the x-axis limit to 10^6
    ax.set_xlim(10, 10**5)  # This sets the limit from 10 to 1,000,000

# Add a big title above all subplots
fig.suptitle('Number of comments per post', fontsize=20)

plt.tight_layout()
plt.show()
```
